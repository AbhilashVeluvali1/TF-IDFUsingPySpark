```
import nltk
from nltk.corpus import stopwords
from nltk.tag import pos_tag
from nltk import ne_chunk
from nltk import word_tokenize
nltk.download('all')
```

```
file = sc.textFile("dbfs:/FileStore/tables/pg1513.txt")
```

```
#collecting all stop words into a set using nltk corpus 
stopWords=set(stopwords.words('english'))
```

```
#cleaning the text to remove all the punctuations and tokenizing it itno words 
import re
cleanText = file.flatMap(lambda x: re.split("[\[\]# _ —./\"'-:;|?*“”’‘]", x)).map(lambda x: x.lower()).filter(lambda x: len(x)>1)
```

```
cleanText.collect()
```

```
#stopwords removal
noStopWords = cleanText.filter(lambda x: x not in stopWords).map(lambda x: x.lower())

noStopWords.collect()
```

```
#Group into named Entities 
namedENGroup = noStopWords.filter(lambda x:pos_tag([x])[0][1] == "NN")
namedENGroup.collect()
```

```
# Frequency of named entities

wordCount=namedENGroup.map(lambda x: (x, 1)).reduceByKey(lambda x, y: x+y).sortBy(lambda x: -x[1])
wordCount.collect()
```
